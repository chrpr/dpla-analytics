```python
import pickle

all_bigrams = pickle.load( open( "/media/storage/dpla-data/words/colls.oct/pickles/all_bfd_top10k.p", "rb" ) )
```

```python
import numpy as np
import pandas as pd

bigrams = []
labels = []
total = []

#" ".join(all_bigrams[:20][0][0])

for bg in all_bigrams[:20]:
    bigrams.append(bg[0])
    labels.append(" ".join(bg[0]))
    total.append(bg[1])

d = {'labels' : pd.Series(labels),
     'total' : pd.Series(total)}

df = pd.DataFrame(d)

df
```

```python
colls = ["artstor","biodiv","rumsey","commonwealth","georgia","harvard",
        "ia","getty","kentucky","minnesota","missouri","mwdl","nara","nocar",
        "smiths","socar","texas","gpo","illinois","usc","virginia","nocoll",
        "hathi","nypl"]

for coll in colls:
    print(coll)
    coll_list = []
    coll_bgs = pickle.load( open( "/media/storage/dpla-data/words/colls.oct/pickles/" + coll + "_bfd.p", "rb" ) )
    for bg in bigrams:
        if bg in coll_bgs: coll_list.append(coll_bgs[bg])
        else: coll_list.append(0)
    df[coll] = pd.Series(coll_list, index=df.index)

df
```

```python
df.to_csv('/media/storage/dpla-data/words/colls.oct/pickles/bigram.new.df.csv')
melted = pd.melt(df, id_vars='labels')
melted.to_csv('/media/storage/dpla-data/words/colls.oct/pickles/bigram.new.melted.csv')
```
